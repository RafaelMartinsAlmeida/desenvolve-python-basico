### Q2.

Dado um número inteiro positivo informado pelo usuário, crie um programa em Python que utilize o comando for para calcular e exibir a soma dos números de 1 até o número informado. Ou seja, $1 + 2 + 3 + ... + n$

Exemplo de interação:
```
Digite um número: 5
A soma dos números de 1 a 5 é 15
```


````python
## Escreva e execute seu código aqui

# Solicita um número inteiro positivo
n = int(input("Digite um número: "))
soma = 0

# Soma todos os valores de 1 até o número informado
for i in range(1, n + 1):
    soma += i

# Exibe o resultado da soma
print(f"A soma dos números de 1 a {n} é {soma}")
````

Escreva um programa que lê 10 valores inteiros positivos do usuário e ao final imprime a média dos valores digitados com duas casas decimais.

```
Digite 10 números positivos:
3
2
12
14
20
42
19
0
8
1
A média dos valores digitados é 12.10
```

````python
## Escreva e execute seu código aqui

# Lê 10 números positivos e calcula a média entre eles

print("Digite 10 números positivos:")
soma = 0

# Lê os 10 valores e acumula a soma
for i in range(10):
    num = int(input())
    soma += num

# Calcula a média
media = soma / 10

# Exibe a média com duas casas decimais
print(f"A média dos valores digitados é {media:.2f}")
````

### Q4.

Você vai criar um sistema que registra os resultados dos jogos do Atlético MG ao longo de um campeonato. Seu sistema vai receber os resultados de todos os jogos do Galo, e deve calcular a pontuação do time sabendo que vitórias valem 3 pontos, empates 1 ponto e derrotas 0 pontos.

Entrada: <br>
A primeira linha de entrada é um inteiro N com a quantidade jogos do galo. Para cada jogo você deve ler 2 inteiros, o primeiro com a quantidade de gols do galo e o segundo com a quantidade de gols do time oponente. 

Saída: <br>
Apresente a soma de vitórias, empates e derrotas do galo, junto com o cálculo da pontuação total.

Ex:
```
Entrada:
4
2
0
1
1
0
3
1
0

Saída:
Vitórias: 2
Empates: 1
Derrotas: 1
Pontuação: 7

```




````python
## Escreva e execute seu código aqui

# Lê a quantidade de jogos e calcula vitórias, empates, derrotas e a pontuação total

N = int(input())  # Número de jogos

vitorias = 0
empates = 0
derrotas = 0

# Para cada jogo, lê os gols do Galo e do adversário
for i in range(N):
    gols_galo = int(input())
    gols_adversario = int(input())

    # Compara os resultados para contar vitórias, empates e derrotas
    if gols_galo > gols_adversario:
        vitorias += 1
    elif gols_galo == gols_adversario:
        empates += 1
    else:
        derrotas += 1

# Calcula a pontuação total
pontos = (vitorias * 3) + empates

# Exibe os resultados
print(f"Vitórias: {vitorias}")
print(f"Empates: {empates}")
print(f"Derrotas: {derrotas}")
print(f"Pontuação: {pontos}")
````

### Q6.

Escreva um programa que lê uma quantidade indefinida de valores e informa o maior e o menor valor digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
12
-1
29
11
-5
0

Saída:
Maior: 29
Menor: -5
```

````python
## Escreva e execute seu código aqui

# Programa que lê números inteiros indefinidamente até o usuário digitar 0.
# Após isso, mostra o maior e o menor número digitado.

maior = None
menor = None

while True:
    valor = int(input())

    if valor == 0:  # Para a leitura se o valor for 0
        break

    # Atualiza maior e menor conforme os valores digitados
    if maior is None or valor > maior:
        maior = valor
    if menor is None or valor < menor:
        menor = valor

print(f"Maior: {maior}")
print(f"Menor: {menor}")
````

### Q7.

Escreva um programa que lê uma quantidade indefinida de valores e informa o **produto** dos valores positivos digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
2
-1
1
3
-5
0

Saída:
Produto: 6
```

````python
## Escreva e execute seu código aqui

# Programa que lê números indefinidamente até o usuário digitar 0.
# Calcula o produto apenas dos números positivos, ignorando os negativos.

produto = 1
tem_positivo = False  # Sinaliza se algum número positivo foi digitado

while True:
    valor = int(input())

    if valor == 0:  # Finaliza o programa ao digitar 0
        break

    if valor <= 0:  # Ignora valores negativos e zero (exceto para encerrar)
        continue

    produto *= valor
    tem_positivo = True

if tem_positivo:
    print(f"Produto: {produto}")
else:
    print("Nenhum valor positivo foi digitado.")
````

### Q8.

Vamos fazer uma calculadora que aceita expressões aritméticas de qualquer tamanho até que o usuário digite a palavra "Fim". Sua calculadora executa apenas as operações "+", "-".

```
Entrada:
5
+
8
-
3
+
2
Fim

Saída:
12
```

````python
## Escreva e execute seu código aqui

# Calculadora simples que realiza soma e subtração de forma contínua.
# O usuário informa o operador (+ ou -) e um número a cada passo.
# Para encerrar, digita a palavra "Fim".

resultado = int(input())  # Primeiro número

while True:
    operador = input()

    if operador == "Fim":  # Encerra o programa
        break

    numero = int(input())

    if operador == "+":
        resultado += numero
    elif operador == "-":
        resultado -= numero
    else:
        print("Operador inválido. Use apenas '+' ou '-'")
        continue

print(resultado)
````
